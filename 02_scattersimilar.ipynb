```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
model=api.load("glove-wiki-gigaword-50")
```

```python

words=["football","hardware","cricket","coding","engineering","coding"]
vectors=[model[word] for word in words]
pca=PCA(n_components=2)
reduce=pca.fit_transform(vectors)
io="football"
similar=model.most_similar(io,topn=2)
print(similar)
#printing with most similar words related word with score
plt.figure(figsize=(12,8))
for i,word in enumerate(words):
    plt.scatter(reduce[i,0],reduce[i,1])
plt.title("PCA visulaization of techonlogy word embeddings")
plt.xlabel("PC1")
plt.ylabel("pc2")
plt.show()
```
