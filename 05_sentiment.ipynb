```python
from transformers import pipeline

print("Enter your choice")
print("1: Sentiment Analysis")
print("2: Text Classification")
print("3: Summarization")
print("4: Translation")
print("5: Question Answering")
print("6: NER")

inp = int(input("Enter your choice: "))

def matched(inp):
    match inp:
        case 1:
            input1 = input("Enter text: ")
            sentiment = pipeline("sentiment-analysis")
            print(sentiment(input1))

        case 2:
            input2 = input("Enter text: ")
            classification = pipeline("text-classification")
            print(classification(input2))

        case 3:
            input3 = input("Enter text to summarize: ")
            summ = pipeline("summarization")
            print(summ(input3))

        case 4:
            input4 = input("Enter text to translate: ")
            translation = pipeline("translation_en_to_fr")  # example
            print(translation(input4))

        case 5:
            question = input("Enter your question: ")
            context = input("Enter context (paragraph): ")
            qna = pipeline("question-answering")
            print(qna(question=question, context=context))

        case 6:
            text = input("Enter text: ")
            ner = pipeline("ner")
            print(ner(text))

        case _:
            print("Invalid choice")

matched(inp)
```
