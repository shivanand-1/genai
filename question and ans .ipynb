```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
loader=PyPDFLoader(r"C:\Users\shiva\Downloads\app.pdf")
document=loader.load() #it will convert into the other into python readable form

text_splitter=RecursiveCharacterTextSplitter(
      chunk_size=100,

       chunk_overlap=20)
chunks=text_splitter.split_documents(document)
print(f"the length of the chunks{len(chunks)}") #it wil overlap the anther words
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=model,
    persist_directory="./chroma_db"
)

# Save to disk
vectorstore.persist()
```
